```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
house=pd.read_csv("C:\\Users\\vp058\\OneDrive\\Desktop\\datasets\\Housing.csv")
house
```

```python
house.describe()
```

```python
house.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
le=LabelEncoder()
```

```python
col=['mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea']
for i in col:
    house[i]=(le.fit_transform(house[i]))

house
```

```python
dummies=pd.get_dummies(house.furnishingstatus)
dummies
```

```python
house=pd.concat([house,dummies],axis="columns")
house
```

```python
house=house.drop("furnishingstatus",axis=1)
house
```

```python
x=house.drop("price",axis=1)
y=house["price"]

print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn import neighbors
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score,r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
scaler=StandardScaler()
```

```python
knn=neighbors.KNeighborsRegressor(n_neighbors=5,weights='distance')
```

```python
pipe_knn=make_pipeline(scaler,knn)
```

```python
pipe_knn.fit(x_train,y_train)
```

```python
y_pred_knn=pipe_knn.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_knn)
mse = mean_squared_error(y_test, y_pred_knn)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred_knn)
print("KNeighborsRegressor Performance:")
print(f"MAE : {mae:,.2f}")
#print(f"MSE : {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
#Applying LINEAR REGRESSION

lr=LinearRegression()
pipe_lr=make_pipeline(scaler,lr)
pipe_lr.fit(x_train,y_train)
y_pred_lr=pipe_lr.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_lr)
mse = mean_squared_error(y_test, y_pred_lr)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred_lr)
print("Linear Regressor's Performance:")
print(f"MAE : {mae:,.2f}")
#print(f"MSE : {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
#Applying RandomForest

rf=RandomForestRegressor(n_estimators=500,random_state=42,max_depth=5)
pipe_rf=make_pipeline(scaler,rf)
pipe_rf.fit(x_train,y_train)
y_pred_rf=pipe_rf.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_rf)
mse = mean_squared_error(y_test, y_pred_rf)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred_rf)
print("Random Forest Regressor's Performance:")
print(f"MAE : {mae:,.2f}")
#print(f"MSE : {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
## Applying Decision tree

dt=DecisionTreeRegressor(random_state=0)
pipe_dt=make_pipeline(scaler,dt)
pipe_dt.fit(x_train,y_train)
y_pred_dt=pipe_dt.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred_dt)
mse = mean_squared_error(y_test, y_pred_dt)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred_dt)
print("Decision Tree's Performance:")
print(f"MAE : {mae:,.2f}")
#print(f"MSE : {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor( n_estimators=100,
        learning_rate=0.05,
        max_depth=5,
        random_state=42)

pipe_xgb=make_pipeline(scaler,xgb)
pipe_xgb.fit(x_train,y_train)
y_pred_xgb=pipe_xgb.predict(x_test)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred_xgb)
mse = mean_squared_error(y_test, y_pred_xgb)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred_xgb)
print("XGBoost Regressor Performance:")
print(f"MAE : {mae:,.2f}")
print(f"MSE : {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

svr=SVR(kernel='linear',epsilon=0.5)

pipe_svr=make_pipeline(scaler,svr)
pipe_svr.fit(x_train,y_train)
y_pred_svr=pipe_svr.predict(x_test)

mae = mean_absolute_error(y_test, y_pred_svr)
mse = mean_squared_error(y_test, y_pred_svr)
rmse = np.sqrt(mse)
r2=r2_score(y_test,y_pred_svr)
print("SVM Regressor Performance:")
print(f"MAE : {mae:,.2f}")
print(f"MSE : {mse:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```
